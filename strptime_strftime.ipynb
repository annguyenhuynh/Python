```python
from datetime import datetime

now = datetime.now()
formatted = now.strftime("%Y-%m-%d %H:%M:%S")
print(formatted)
```

* string format time (strftime): format date time into readable strings

* strptime (string parse time) does the opposite of strftime — it parses a string into a datetime object using a specified format.

```python
date_string = "2025-03-08 22:15:00"
dt = datetime.strptime(date_string, "%Y-%m-%d %H:%M:%S")

print(dt)
```

#### Common **strptime** use cases

##### 1. Parse a date only

```python
datetime.strptime("2025-03-08", "%Y-%m-%d")
```

##### 2. Parse with month name

```python
datetime.strptime("08 March 2025", "%d %B %Y")
```

##### 3. Parse with abbreviated month

```python
datetime.strptime("08 Mar 2025", "%d %b %Y")
```

##### 4. Parse mm/dd/yyyy

```python
datetime.strptime("03/08/2025", "%m/%d/%Y")
```

```python
dt = datetime.strptime("03/08/25", "%m/%d/%y")
formatted = dt.strftime("%Y-%m-%d")

formatted
```

| Code | Meaning             |
| ---- | ------------------- |
| `%Y` | 4-digit year        |
| `%y` | 2-digit year        |
| `%m` | month (01–12)       |
| `%d` | day of month        |
| `%H` | hour (00–23)        |
| `%M` | minute              |
| `%S` | second              |
| `%b` | abbreviated month   |
| `%B` | full month name     |
| `%a` | abbreviated weekday |
| `%A` | full weekday        |
