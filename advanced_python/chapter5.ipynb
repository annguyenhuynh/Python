```python
# Atomic numbers of noble gases
nobles = {'He':2, 'Ne':10, 'Ar':18, 'Kr':36, 'Xe':54}
def show_element(elements):
    for element in elements:
        print('Atomic number of {} is {}'.format(
            element,nobles[element]
        ))
```

```python
try:
    show_element(['Ne', 'Ar', 'Br'])
except Exception as e:
    missing_element = e.args[0]
    print(f"Missing element: {missing_element}")
```

### Raising Exceptions

* **ValueError** is a built-in exception that signals some data is of the correct type, but the format isn't valid

```python
import re
class Money:
    def __init__(self,dollars,cents):
        self.dollars = dollars
        self.cents = cents
    
    def __repr__(self):
        return f"Money({self.dollars}, {self.cents})"

    def money_from_string(amount):
        # amount is a string like "140.75"
        match = re.search(
            r'^\$(?P<dollars>\d+)\.(?P<cents>\d\d)$', amount)
        if match is None:
            raise ValueError ("Invalid amount: {amount}")
        dollars = int(match.group('dollars'))
        cents = int(match.group('cents'))
        return Money(dollars,cents)
```

```python
money_from_string('$Big Money')
```

```python
money_from_string("$400.24")
```
