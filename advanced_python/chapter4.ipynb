### Basic Decorators

```python
def printlog(func):
    def wrapper(arg):
        print("CALLING:  " + func.__name__)
        return func(arg)
    return wrapper
```

```python
@printlog 
def foo(x):
    print(x+2)
```

```python
foo(3)
```

### Generic Decorators

* Decorated functions should be generic

```python
## Better version of @printlog

def print_log(func):
    def wrapper_1(*args, **kwargs):
        print("CALLING " + func.__name__)
        return func(*args, **kwargs)
    return wrapper_1
```

```python
@print_log
def bar (x,y):
    return x*y

bar(2,3)
```

**Note**
* When writing decorators for methods, it's recommended to have wrappers take ***args** and ****kwargs**

### Data in Decorators

```python
"""
Imagine you need to keep running average of what a chosen function runs. 
And further, you need to run this for a family of functions or methods. 
We can write a decorator called running_average to handle this
"""
```

```python
def running_average(func):
    data = {"total":0, "count":0}
    def wrapper(*args, **kwargs):
        val = func(*args, **kwargs)
        data["total"] += val
        data["count"] +=1
        print("Average of {} so far: {:.01f}".format(
            func.__name__, data["total"]/data["count"]
        ))
        return val 
    return wrapper
```

```python
@running_average
def foo(x):
    return x+2
```

```python
foo(1)
```

```python
foo(2)
```

```python
foo(5)
```

```python
foo(7)
```

```python
data = {"total":0, "count":0}
def outside_data_running_average(func):
    def wrapper(*args, **kwargs):
        val = func(*args, **kwargs)
        data["total"] += val
        data["count"] +=1
        print("Average of {} so far: {:.01f}".format(
            func.__name__, data["total"]/data["count"]
        ))
        return val 
    return wrapper
```

```python
@outside_data_running_average
def foo(x):
    return x+2
```

```python
foo(1)
```

```python
foo(2)
```

```python
@outside_data_running_average
def bar(x):
    return 3*x
```

```python
bar(10)
```

```python
foo(1)
```

### Accessing Inner Data

```python
def collectstats(func):
    data = {"total":0, "count":0}
    def wrapper(*args, **kwargs):
        val = func(*args, **kwargs)
        data["total"] += val
        data["count"] +=1
        print("Average of {} so far: {:.01f}".format(
            func.__name__, data["total"]/data["count"]
        ))
        return val 
    wrapper.data=data
    return wrapper
```

```python
@collectstats
def foo(x):
    return x+2
```

```python
foo.data
```

```python
foo(1)
foo.data
```

```python
foo(2)
```

```python
foo.data
```
