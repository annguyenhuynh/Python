```python
import datetime
import pytz
```

```python
# Simple datetime
d = datetime.date(2020, 6 ,12)
print(d)
```

```python
tday = datetime.date.today()
print(tday.weekday())       # Monday is 0 and Sunday is 6
print(tday.isoweekday())    # Monday is 1 and Sunday is 7
```

```python
# Timedeltas -- difference in 2 dates or times
tdelta = datetime.timedelta(days=7)
print(tday + tdelta)
```

```python
dt = datetime.datetime(2020, 4, 12, 9, 45, 55, 100000)
print(dt)
print(dt.date())
print(dt.time())
```

```python
dt_today = datetime.datetime.today()
dt_now = datetime.datetime.now()
dt_utcnow = datetime.datetime.utcnow()

print(dt_today)
print(dt_now)
print(dt_utcnow) #UTC is 5 hours ahead of EST 
```

```python
dt = datetime.datetime(2015,6,25,12,30,45,tzinfo=pytz.UTC)
print(dt)

dt_utcnow = datetime.datetime.now(tz=pytz.UTC)
print(dt_utcnow)
```

```python
dt_est = datetime.datetime.now() # naiive dateime wihtout timezone aware
est_tz = pytz.timezone('US/Eastern')

dt_est = est_tz.localize(dt_est)

dt_mnt = dt_est.astimezone(pytz.timezone('US/Mountain'))
print(dt_mnt.isoformat())

print(dt_mnt.strftime('%d %B, %Y'))
```
