```python
import threading 
import time
```

```python
def sleeper(n, name):
    print("Hi, I am {}, I'm going to sleep for 5 seconds\n".format(name))
    time.sleep(n)
    print("{} has woken up from sleep \n".format(name))

t = threading.Thread(
    target=sleeper, 
    name="thread1",
    args=(5, "thread1") #arguments for the target function
)

t.start() #start the thread

# Wait for the thread to finish
t.join()

## Concurrency
print('hello')
print('hello')
```
