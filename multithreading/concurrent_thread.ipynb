```python
import threading 
import time
```

```python
def sleeper(n, name):
    print("Hi, I am {}, I'm going to sleep for 5 seconds\n".format(name))
    time.sleep(n)
    print("{} has woken up from sleep \n".format(name))

t = threading.Thread(
    target=sleeper, 
    name="thread1",
    args=(5, "thread1") #arguments for the target function
)


t.start()

thread_list = []

start = time.time()
for i in range(5):
    t = threading.Thread(
        target = sleeper,
        name = 'thread{}'.format(i),
        args=(5, 'thread{}'.format(i))
    )

    thread_list.append(t)  # Add the thread to the list

    t.start()

    print("{} has started\n".format(t.name))

# Ensure that other programs cannot access main program threads until all threads have completed
for t in thread_list:
    t.join()

end = time.time()
print('time taken: {}'.format(end-start))
print("All 5 threads have completed")
```
