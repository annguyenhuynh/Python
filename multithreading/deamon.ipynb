```python
import time
import threading
from threading import Thread
```

```python
total = 4
lock = threading.Lock()

def create_items():
    global total

    for i in range (10):
        time.sleep(2)
        print("added item")
        total += 1
        print("creation is done")

def create_items_2():
    global total

    for i in range (7):
        time.sleep(1)
        print("added item")
        total += 1
        print("creation is done")

def limit_item():
    global total
    while True:
        with lock:
            if total > 5:
                print("Overload")
                total -= 3
                print("Subtracted 3 items")
            else:
                print("waiting")
        time.sleep(1)

creator_1 = threading.Thread(target=create_items)
creator_2 = threading.Thread(target=create_items_2)
limitor = threading.Thread(target=limit_item, daemon=True)

creator_1.start()
creator_2.start()
limitor.start()

creator_1.join()
creator_2.join()


print(f"The ending of total is, {total}")
```
