### 1. List Comprehension

```python
values = []
for x in range(10):
    values.append(x)

# List comprehension
values = [x for x in range(10)]
```

```python
# Get all even numbers from 0 - 50

even = []

for x in range(50):
    if x % 2 == 0:
        even.append(x)
    else:
        print(f"{x} is not an event number")
```

```python
# Doing the same thing above using list comprehension
evens = [x for x in range(50) if x%2==0]
print(evens)
```

```python
# More complex example
# String that start with letter a and end with letter y
options = ["all", "albany", "apple", "world", "hello", ""]
valid_strings = [
    string for string in options 
    if len(string) >=2
    if string[0] == "a"
    if string[-1] == "y"
]

print(valid_strings)
```

#### Nested List Comprehension

```python
matrix = [[1,2,3], [4,5,6], [7,8,9]]
flatten = []

for row in matrix:
    for num in row:
        flatten.append(num)
```

```python
flattened = [num for row in matrix for num in row] #value to return -> outer for then inner for
print(flattened)
```

## 2. If/Else in Comprehension

```python
categories = []

for number in range(10):
    if number % 2 == 0:
        categories.append("Even")
    else:
        categories.append("Odd") 

print(categories)
```

```python
# List comprehensionn for above problem
categories = ["Even" if x % 2 == 0 else "Odd" for x in range(10)]
```

#### Nested 3-dimensional list

```python
# Building a 3D-list
import pprint 
printer = pprint.PrettyPrinter()

lst = []

for a in range(5):
    l1=[]
    for b in range(5):
        l2 = []
        for num in range(5):
            l2.append(num)
        l1.append(l2)
    lst.append(l1) 

printer.pprint(lst)
```

```python
# List comprehension to create 3D list
lst = [[[num for num in range(5)] for _ in range(5)] for _ in range(5)]
printer.pprint(lst)
```

## 3. Transformation in Comprehension

```python
# Work if we have complicated function
def sqare(x):
    return x**2 

squares = [sqare(x) for x in range(10)]
```

## 4. Dictionary Comprehension

```python
# Creating a dictionary
pairs = [("a",1), ("b",2), ("c",3)]

my_dict = {k: sqare(v) for k, v in pairs}
print(my_dict)
```

## 5. Set Comprehension

```python
# Remove duplicates from a list while applying a function
nums = [1,2,2,3,3,3,4,4,4,4]

unique_sequences = {x**2 for x in nums}
print(unique_sequences)
```

## 6. Generator Comprehension

* A generator comprehension (or generator expression) in Python is a concise way to create a generator â€” an iterator that yields items one by one lazily (only when needed), instead of building the entire list or set in memory.

* It looks similar to a list comprehension, but uses parentheses () instead of square brackets [].

```python
sum_of_squares = sum(x**2 for x in range(10000))
```

```python
print(sum_of_squares)
```
