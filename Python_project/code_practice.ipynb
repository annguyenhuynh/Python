```python
# leap year
def leap_year(year):
    leap = False
    if year % 4 == 0:
        if year % 100 == 0:
            if year % 400 == 0:
                leap = True
        else:
            leap = True
    return leap
```

```python
# Find the second 
n = int(input())

students = []

for _ in range(n):
    name = input()
    scores = float(input())
    students.append([name,scores])

# Generate unique scores
unique_scores = sorted({s for _, s in students})

# Get the second lowest scores 
second_score = unique_scores[1]

# Print out the name of the person of the second lowest score
second_score_name = sorted([name for name, s in students if s == second_score ])

for name in second_score_name:
    print(name)
                           
# The reson we use placeholder is because we tell Python tp “Run this loop a certain number of times (whatever number the user enters),
# but I don’t actually need to use the loop variable.”
```

* Step-by-step breakdown:

**for _, s in students**

students is a list of lists (nested list) — e.g.

students = [["Harry", 37.21], ["Berry", 37.21], ["Tina", 37.2]]


* When you write for _, s in students, each sublist [name, score] is unpacked into two variables:

_ → gets the name (we don’t use it, so _ is a placeholder)

s → gets the score

{s for _, s in students}

* This is a set comprehension — it builds a set of all the s (scores).

* The curly braces {} mean "make a set" (unique values only).

```python
# Print out the average score of each student in the query. Scores are stored as list of values

n = int(input())
students = {}

for _ in range(n):
    data = input().split()
    name = data[0]
    scores = list(map(float, data[1:]))
    students[name] = scores 

    query_name = input().strip()

    avg_score = sum(students[query_name])/len(students[query_name])

    print(f"{avg_score:.2f}")
```

### map, zip, enumerate

#### 1. map()

* allow you to pass a function to all elements in an iterable

```python
lst = [1000, 4000, 16000, 150, 9001, 1500]

def greater_than_9000(x):
    if x > 9000:
        return True
    return False 

result = map(greater_than_9000, lst)
print(result)
```

```python
# Extract mapping into a new list
mapped_list = [*map(greater_than_9000,lst)]
print(mapped_list)
```

```python
# Use map with lambda function
mapped_list_1 = [*map(lambda x:x>9000, lst)]
print(mapped_list_1)
```

| Concept                | What it is                                                             | Purpose                                       |
| ---------------------- | ---------------------------------------------------------------------- | --------------------------------------------- |
| **List comprehension** | A **compact syntax** to build a new list from an existing iterable     | To **generate** or **filter** data            |
| **Lambda function**    | A **small anonymous function** (no name) created with `lambda` keyword | To **define a function inline** for short use |


| Feature                       | **List Comprehension**                | **Lambda Function**                                                    |
| ----------------------------- | ------------------------------------- | ---------------------------------------------------------------------- |
| **Creates**                   | A list                                | A function                                                             |
| **Returns**                   | A list object                         | A function object                                                      |
| **Common Use**                | Generating lists quickly              | Writing short, inline functions (e.g., with `map`, `filter`, `sorted`) |
| **Needs loop?**               | Yes, usually iterates over items      | No, applies logic to inputs                                            |
| **Can include conditionals?** | Yes (`if` inside comprehension)       | Yes (inside expression)                                                |
| **Example use**               | `[x for x in range(5) if x % 2 == 0]` | `lambda x: x % 2 == 0`                                                 |


##### Lambda function

```python
add_on = lambda x, y: x+y

result = add_on(1,7)
print(result)
```

```python
# map function with lambda function
my_nums = [1,2,3,4,5,6,7,8,9]
resp = list(map(lambda x: x**2,my_nums))
print(resp)
```

```python
# filter function with lambda function
evens = list(filter(lambda x: x%2==0, my_nums))
print(evens)
```

```python
# sort function with lambda function
values = [(1, "b", "hello"),(2, "a", "world"), (3, "c", "OK")]

sorted_values = sorted(values, key=lambda x: x[1])
print(list(sorted_values))
```

```python
# More advanced examples

from functools import reduce

numbers = [1,2,3,4,5]

#Using reduce to sum the list without initializer
sum_of_numbers = reduce(lambda acc, x:acc+x, numbers)
print(sum_of_numbers)

#Using reduce to find the maximum value
max_value = reduce(lambda acc, x:acc if acc > x else x, numbers)
print(max_value)
```

#### 2. zip function
* The built-in zip() function in Python is used to combine multiple iterables (like lists, tuples, etc.) element-wise into tuples. It “zips” them together so that the first elements form one tuple, the second elements form another, and so on.

```python
latitude = [4,5,6,4,6,2,4,5,6]
longitude = [6,3,6,3,4,5,6,8,5]

[*zip(latitude, longitude)] #<-- More pythonic

# The '*' operator is used to unpack the zipped object into a list of tuples. (like list(zip(latitude, longitude)))
```

```python
# zip will continue up to the length of the shortest input iterable.

short = [1,2,3,4]
long = [1,2,3,4,5,6,7,8]

[*zip(short, long)] 
```

```python
# If you want to keep all items, user itertools.zip_longest

from itertools import zip_longest
[*zip_longest(short, long, fillvalue=0)]
```

#### 3. Enumerate
* return sequence of (index, item) tuples

```python
characters = ["Krillin", "Goku", "Vegeta", "Gohan", "Piccolo"]
list(enumerate(characters))
```

```python
# Use enumerate() in a for loop to get an index and its item

for index, character in enumerate(characters):
    print(index, character)
```

```python
# Store index of duplicate items in a dictionary
characters_1 = ["Krillin", "Goku", "Goku", "Gohan", "Piccolo",
              "Krillin", "Goku", "Vegeta", "Gohan", "Piccolo",
              "Piccolo", "Goku", "Vegeta", "Goku", "Piccolo"]

character_map = {character:[] for character in set(characters)}


for index, character in enumerate(characters_1):
    character_map[character].append(index)

character_map
```
